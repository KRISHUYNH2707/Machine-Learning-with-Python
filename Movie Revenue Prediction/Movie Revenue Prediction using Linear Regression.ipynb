## Gather data 

* Data on feature -> movie bugest in USD
* Data on target -> movie revenue in USD

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns=["production_budget_usd"])
y = pd.DataFrame(data, columns=["worldwide_gross_usd"])
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title("Film Cost vs Global Budget")
plt.xlabel("Product Budget $")
plt.ylabel("Worldwide Gross $ ")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
```

```python
regression.fit(X,y)
```

```python
slope = regression.coef_ # Slope coeffiecient
```

```python
intercept = regression.intercept_ #intercept coefficient
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=4)

plt.title("Film Cost vs Global Budget")
plt.xlabel("Product Budget $")
plt.ylabel("Worldwide Gross $ ")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```
