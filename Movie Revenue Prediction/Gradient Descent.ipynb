# Example 1 

## $ f(x) = x^2 + x + 1 $

```python
def f(x):
    return x** 2 + x + 1 
```

```python
# Generate data
x_data = np.linspace(start=-3,stop=3, num=100)
```

```python
plt.figure(figsize=[10,5])

#cost function chart
plt.subplot(1,2,1)

plt.plot(x_data, f(x_data), color='blue', linewidth=3)
plt.xlabel('X')
plt.ylabel("f(x)")
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title("Cost function", fontsize=17)

#derivate chart
plt.subplot(1,2,2)
plt.plot(x_data, df(x_data), color='skyblue',linewidth="5")
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel("X")
plt.ylabel("df(x)")

plt.show()
```

## Slopes and Derivatives

```python
def df(x):
    return 2*x + 1 
```

```python
new_x = 3
previous_x = 0
learning_rate = 0.1 
precision = 0.001

x_list = []
slope_list = []

for i in range(500):
    previous_x = new_x 
    gradient = df(previous_x)
    new_x = previous_x - learning_rate * gradient
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print(i)
        break 

print("Local minimum is at:", new_x)
print("Slope at the minimum point is: ", df(new_x))
print("Value at the minimum point is: ", f(new_x))
```

```python
plt.figure(figsize=[20,5])

#cost function chart
plt.subplot(1,3,1)

plt.plot(x_data, f(x_data), color='blue', linewidth=3)

values = np.array(x_list)
plt.scatter(x_list, f(values), color="red", s=100, alpha=0.6)

plt.xlabel('X')
plt.ylabel("f(x)")
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title("Cost function", fontsize=17)

#derivate chart
plt.subplot(1,3,2)
plt.plot(x_data, df(x_data), color='skyblue',linewidth="5")

slope_values = np.array(slope_list)
plt.scatter(slope_list, df(slope_values), color="red", s=100, alpha=0.6)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel("X")
plt.ylabel("df(x)")


#derivate chart
plt.subplot(1,3,3)
plt.plot(x_data, df(x_data), color='skyblue',linewidth="5")

plt.scatter(slope_list, df(slope_values), color="red", s=300, alpha=0.6)
plt.grid()
plt.xlim(-0.5,1)
plt.ylim(0.8,3.2)
plt.title('Slope of the cost function (Close up)', fontsize=17)
plt.xlabel("X")
plt.ylabel("df(x)")

plt.show()
```
